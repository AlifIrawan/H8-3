### Data Visualization using PCA



```python
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x) # Normalizing the features
```

```python
x.shape
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
normalised_breast = pd.DataFrame(x, columns=feat_cols)
normalised_breast.tail()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_DF = pd.DataFrame(data=principalComponents_breast, columns=['Principal Component 1', 'Principal Component 2'])
```

```python
principal_breast_DF.tail()
```

```python
print('Explained Variation per principal component : {}'.format(pca_breast.explained_variance_ratio_))
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel("Principal Component - 1", fontsize=20)
plt.ylabel("Principal Component - 2", fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset", fontsize=20)

targets = ['Benign', 'Maligant']
colors = ['r', 'g']
for target, color, in zip(targets, colors):
    indicesToKeep = breast_dataset['label'] == target
    plt.scatter(principal_breast_DF.loc[indicesToKeep, 'principal component 1'], principal_breast_DF.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)
    
plt.legend(targets, prop={'size': 15})

plt.show()
```

## How is Clustering an Unsupervised Learning Problem?

Misalkan kita sedang mengerjakan sebuah proyek dimana kita perlu memprediksi penjualan pasar besar :



Atau proyek dimana tugas kita adalah memprediksi apakah pinjaman akan disetujui atau tidak :



## Property of Clusters











### Property 1

**All the data points in a cluster should be similar to each other**. Mari kita mengilustrasikannya menggunakan contoh diatas :





### Property 2

**The data points from different clusters should be as different as possible**. Ini secara intuitif akan masuk akal jika kita memahami properti diatas. Mari kita ambil contoh yang sama untuk memahami properti ini :



**Manakah dari kasus berikut yang menurut kalian akan memberi kita cluster yang lebih baik?** 

Jika melihat kasus 1 :



Pelanggan dalam kelompok merah dan biru sangat mirip satu sama lain. Empat poin teratas di cluster merah memiliki properti yang serupa dengan 2 pelanggan teratas di cluster biru. Mereka memiliki pendapatan tinggi dan nilai hutang yang tinggi. Disini kita mengelompokkan mereka secara berbeda. 

Sedangkan jika melihat kasus ke 2 :



Poin di Cluster merah sama sekali berbeda dari pelanggan di cluster biru. Semua pelanggan cluster merah memiliki pendapatan tinggi dan hutang tinggi sedangkan pelanggan di cluster biru memiliki pendapatan tinggi dan hutang rendah. Jelas kita memiliki clustering pelanggan yang lebih baik dalam kasus ini.

Oleh Karena itu, Poin data dari cluster yang berbeda harus sama berbeda satu sama lain mungkin untuk memiliki cluster yang lebih bermakna.

## Understanding the Different Evaluation Metrics for Clustering



### Inertia



#### Step 1 : Choose the number of clusters *k*

Langkah pertama dalam k-means adalah memilih jumlah cluster k

#### Step 2 : Select *k* random points from the data as centroids

Selanjutnya, kita secara acak memilih centroiduntuk setiap cluster. Misalkan kita ingin memiliki 2 cluster, jadi *k* sama dengan 2 disini. Kita kemudian secara acak memilih centroid :



Disini, Lingkaran merah dan hijau mewakili centroid untuk kelompok ini.

#### Step 3 : Assign all the points to the closest cluster centroid

Setelah kita menginisialisasi centroids, kita menetapkan setiap titik ke cluster centroid terdekat :



Disini kita dapat melihat bahwa titik-titik yang lebih dekat ke titik merah ditetapkan ke kelompok merah sedangkan titik-titik yang lebih dekat ke titik hijau ditetapkan ke kelompok hijau.

#### Step 4 :Recompute the centroids of newly formed clusters

Sekarang setelah kita menetapkan semua titik ke salah satu cluster, langkah selanjutnya adalah menghitung centroid dari cluster yang baru terbentuk dengan menggeser centroid ke titik tengah dari data :



Disini, Silang merah dan hijau adalah centroids baru

#### Step 5 :Repeat steps 3 and 4

Kita kemudian mengulangi langkah 3 dan 4 :



Langkah menghitung centroid dan menetapkan semua titik ke kluster berdasarkan jaraknya dari centroid adalah satu iterasi. Tapi tunggu kapan kita harus menghentikan proses ini?

## K-Means with Python

```python
from sklearn.cluster import KMeans
```

```python
X = -2 * np.random.rand(100, 2)
X1 = 1 + 2 * np.random.rand(50, 2)
X[50:100, :] = X1

plt.scatter(X[ : , 0], X[ : , 1], s=50, c='b')
plt.show()
```

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11), wcss) 
plt.title("The Elbow Method")
plt.xlabel("Number of Cluster")
plt.ylabel("WCSS") #Within cluster sun of squares
plt.show()
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv')
data.head()
```

```python
data.describe()
```

```python
# Standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Statistics of scaled data
pd.DataFrame(data_scaled).describe()
```

```python
# Defining the Kmeans function with initialization as K-means++
kmeans = KMeans(n_clusters=2, init='k-means++')

# Fitting the k-means algorithm on scaled data
kmeans.fit(data_scaled)
```

```python
# Inertia on the fitted data
kmeans.inertia_
```

```python
# Fitting multiple k-means algorithms and storing the valuees in an empty list
SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_clusters=cluster, init='k-means++')
    kmeans.fit(data_scaled)
    SSE.append(kmeans.inertia_)
    
# Converting the results into a dataframe and plotting them
frame = pd.DataFrame({'Cluster': range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
```

# Save and Load Models

## Finalize Your Model with Pickle



```python
# Assigning features and label variables

# First Feature
weather = ['Sunny', 
           'Sunny', 
           'Overcast', 
           'Rainy', 
           'Rainy', 
           'Overcast', 
           'Sunny', 
           'Sunny', 
           'Rainy', 
           'Sunny', 
           'Overcast', 
           'Overcast', 
           'Rainy']

# Second Feature
temp = ['Hot', 
        'Hot', 
        'Hot', 
        'Mild',
        'Cool', 
        'Cool', 
        'Cool', 
        'Mild', 
        'Cool', 
        'Mild', 
        'Mild', 
        'Mild', 
        'Hot']

# Label or Target varible
play = ['No', 
        'No', 
        'Yes', 
        'Yes', 
        'Yes', 
        'No', 
        'Yes', 
        'No', 
        'Yes', 
        'Yes', 
        'Yes', 
        'Yes', 
        'Yes']

# Import labelEndcoder
from sklearn import preprocessing

# Creating labelendcoder
le = preprocessing.LabelEncoder()

# Converting weather
weather_endcoded = le.fit_transform(weather)

# Converting temp
temp_endcoded = le.fit_transform(temp)

# Converting play
label = le.fit_transform(play)

# Combining weather and temp into single listof tuple
features = list(zip(weather_endcoded, temp_endcoded))

# Import Gaussian naive bayes model
from sklearn.naive_bayes import GaussianNB

# Create a Gaussian Classifier
model = GaussianNB()

# Train the model using the training sets
model.fit(features, label)

# Save the model to disk
import pickle

filename = 'finalized_model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open('finalized_model.sav', 'rb'))
```

```python
loaded_model.predict([[0, 2]]) # 0: Overcast, 2: Mild
```

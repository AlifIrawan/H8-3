# Regression
Adalah salah satu bidang terpenting dalam statistik dan machine learning. Gampang-nya Regression mencari hubungan antar variable. Regression adalah bentuk lain dari supervised learning.

Perbedaan antara Classification dan Regression adalah bahwa regression **menghasilkan angka** daripada kelas. Regression berguna saat memprediksi masalah berbasis angka seperti **Harga pasar saham**, **Suhu** dan **Probabilitas suatu peristiwa**.

**Misalnya**, kita dapat mengamati beberapa karyawan dari beberapa perusahaan dan mencoba memahami bagaimana gaji mereka bergantung pada fitur, seperti pengalaman, tingkat pendidikan, peran, kota, dan sebagaiannya.

Contoh diatas adalah **masalah regresi** dimana data yang terkait dengan setiap karyawan merupakan satu observasi. Asumsinya pengalaman, pendidikan, peran, dan kota adalah ciri ciri tersendiri, sedangkan gaji bergantung pada-nya. **Sama hal-nya** dengan contoh real-estate tadi.

Masalah Regression biasanya memiliki 1 variable depeden yang kontinu dan tidak terbatas, namun masukan dapat berupa data kontinu, diskrit atau bahkan kategorikal seperti jenis kelamin, kebansaan, merek dan lain lain.

**When do you need Regression?**
Biasanya, kita memerlukan regresi untuk menjawab apakag dan bagaimana beberapa fenomena memengaruhi yang lain atau bagaimana terkaitnya variable yang ada. Misalnya, kita dapat menggunakannya untuk menentukan apakah dan sejauh mana pengalaman atau gender mempengaruhi gaji.

Regression juga berguna saat kita ingin meramalkan respons menggunakan kumpulan prediktor baru. Misalnya, kita dapat mencoba memprediksi konsumsi listrik rumah tetangga dalam 1 jam kedepan dengan melihat suhu luar ruangan, waktu dan jumlah orang dalam rumah tersebut.

# Implementing Linear Regression in Python


## Simple Linear Regression with Scikit-Learn


```python
# Step 1 : Import Packages and Classes
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
x = np.array([5,15,25,35,45,55]).reshape((-1,1))
y = np.array([5,20,14,32,22,38])
```

```python
print(x)
print(y)
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
```

```python
model = LinearRegression().fit(x, y)
```

```python
r_sq = model.score(x, y)
print('coefficient of determination : ', r_sq)
```

```python
print('Intercept : ', model.intercept_)
print('Slope :', model.coef_)
```

```python
new_model = LinearRegression().fit(x, y.reshape((-1, 1)))
print('Intercept : ', new_model.intercept_)
print('Slope :', new_model.coef_)
```

```python
y_pred = model.predict(x)
print('Predicted response : ', y_pred, sep='\n')
```

```python
plt.scatter(x, y, alpha=.5)
plt.plot(x, y_pred)

plt.title('Scatter plot x dan y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
y_pred = model.intercept_ + model.coef_ * x
print('predicted response : ', y_pred, sep='\n')
```

```python
x_new = np.arange(5).reshape((-1, 1))
print(x_new)
y_new = model.predict(x_new)
print(y_new)
```

## Multiple Linear Regression with Scikit-Learn

```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = [[0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]]
y = [4, 5, 20, 14, 32, 22, 38, 43]
x, y = np.array(x), np.array(y)
```

```python
print(x)
print(y)
```

```python
model = LinearRegression().fit(x, y)
```

```python
r_sq = model.score(x, y)
print('coefficient of determination', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

```python
y_pred = model.predict(x)
print('predicted response : ', y_pred, sep='\n')
```

Respons yang diprediksi yang diperoleh dengan .predict() sangat mirip dengan berikut ini :

```python
y_pred = model.intercept_ + np.sum(model.coef_ * x, axis=1)
print('predicted response : ', y_pred, sep='\n')
```

```python
x_new = np.arange(10).reshape((-1, 2))
print(x_new)

y_new = model.predict(x_new)
print(y_new)
```

```python
x_ = PolynomialFeatures(degree=2, include_bias=False).fit_transform(x)
```

```python
print(x_)
```

```python
model = LinearRegression().fit(x_, y)
```

```python
r_sq = model.score(x_, y)
print('coefficient of determination', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

```python
x_ = PolynomialFeatures(degree=2, include_bias=True).fit_transform(x)
```

```python
print(x_)
```

```python
model = LinearRegression(fit_intercept=False).fit(x_, y)
```

```python
r_sq = model.score(x_, y)
print('coefficient of determination', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

```python
y_pred = model.predict(x_)
print('Predicted response', y_pred, sep='\n')
```

```python
# Step 1 : Import Packages
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Step 2a : Provide data
x = [[0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]]
y = [4, 5, 20, 14, 32, 22, 38, 43]
x, y = np.array(x), np.array(y)

# Step 2b : Transform input data
x_ = PolynomialFeatures(degree=2, include_bias=False).fit_transform(x)

# Step 3 : Create a model and fit it
model = LinearRegression().fit(x_, y)

# Step 4 : Get results
r_sq = model.score(x_, y)
intercept, coefficients = model.intercept_, model.coef_

# Step 5 : Predict
y_pred = model.predict(x_)
```

```python
print('Coefficient of Determination :', r_sq)
print('Intercept :', intercept)
print('Coefficients :', coefficients, sep='\n')
print('Predicted response :', y_pred, sep='\n')
```

## Advanced Linear Regression with Statsmodel

```python
import numpy as np
import statsmodels.api as sm
```

```python
x = [[0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]]
y = [4, 5, 20, 14, 32, 22, 38, 43]
x, y = np.array(x), np.array(y)
print(x)
```

```python
x = sm.add_constant(x)
```

```python
print(x)
print(y)
```

```python
model = sm.OLS(y, x)
```

```python
results = model.fit()
```

```python
print(results.summary())
```

```python
print('Coefficient of Determination :', results.rsquared)
print('Adjusted Coefficient of Determination :', results.rsquared_adj)
print('Regression Coefficient :', results.params)
```

```python
print('Predicted Response : ', results.fittedvalues, sep='\n')
print('Predicted Response : ', results.predict(x), sep='\n')
```

```python
x_new = sm.add_constant(np.arange(10).reshape((-1, 2)))
y_new = results.predict(x_new)

print(x_new)
print(y_new)
```

## Using SKlearn Example 2 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/CarPrice_Assignment.csv', usecols=['horsepower', 'price'])
df.head()
```

```python
df.shape
```

```python
f = plt.figure(figsize=(12, 4))

f.add_subplot(1,2,1)
df['horsepower'].plot(kind='kde')

f.add_subplot(1,2,2)
plt.boxplot(df['horsepower'])

plt.show()
```

```python
plt.scatter(df['horsepower'], df['price'])

plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.title('Scatter Plot Horsepower vs Price')

plt.show()
```

```python
df.corr()
```

```python
x = df['horsepower'].values.reshape(-1,1)
y = df['price'].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
```

```python
LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
lin_reg.score(x_test, y_test)
```

```python
y_prediction = lin_reg.predict(x_test)

plt.scatter(x_test, y_test)
plt.plot(x_test, y_prediction, c='r')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.title('Plot Horsepower Vs Price')
```

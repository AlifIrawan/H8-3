## Machine Learning
adalah penerapan dari artificial intelligence (AI) yang memberikan kemampuan kepada komputer untuk belajar dan berkembang dari pengalaman tanpa harus diprogram secara explicit.

Ada berbagai algoritma Machine Learning seperti :
 - Decision Trees
 - Naive Bayes
 - Random Forest
 - Support Vector Machine
 - K-nearest neighbor
 - K-means Clustering
 - Gaussian Mixture Model
 - Hidden Markov Model
 - etc.
 

**Example** : Misal kita harus memprediksi harga rumah, dengan kumpulan data yang terdiri dari biaya rumah dan jumlah kamar di rumah, serta 1000 rumah dengan atribut serupa. Baik harga dan jumlah kamar adalah fitur. Sekarang tujuan kita adalah memasukkan kedua fitur ini ke algoritma. Input kita, dalam hal ini adalah jumlah kamar dan algoritma harus memprediksi harga rumah. Algoritma akan mencoba mempelajari hubungan antara jumlah kamar dan biaya rumah.

### Terminology

- **Dataset** : A set data examples
- **Features** : Important pieces of data that help us understand a problem. These are fed in to a Machine Learning algorithm to help it learn.
- **Model** : The representation (Internal Model) of a phenomenon that a Machine Learning algorithm has learnt. it learns this from the data it is shown during training. The model is the output you get after training an algorithm. For example Desicison Tree Algorithm would be trained and produce a decision tree model.

### Process

1. **Data Collection** : Collect the data that the algortihm will learn from.
2. **Data Preparation** : Format and engineer the data into the optimal format.
3. **Training** : Also known as the fitting stage, this is where the Machine Learning algorithm actually learns by showing it the data that has beed collected and prepared
4. **Evaluation** : Test the model to see how well its perform's
5. **Tuning** : Fine tune the model to maximise it's performance

### Supervised Learning
Merupakan suatu teknik pembelajaran yang didalam keseluruhan pembelajarannya **di-bimbing oleh supervisi**. **Tujuan utama-nya** adalah untuk memprediksi hasil yang diberikan serangkaian training samples bersama dengan training labels, yang juga dikenalkan sebagai **Classifying** data point.Karena kita memberi tahu algoritma pada saat pelatihan, apa yang diprediksi, Karennya hal itu disebut **Supervised Learning**.

**Misalkan**, Kita memiliki 10.000 gambar, 5.000 gambar kucing dan 5.000 gambar anjing pada masing masing gambar memiliki label nol dan satu. Tujuan dari teknik ini adalah untuk menemukan pola dalam data yang diberi batasan sebagai label. Algoritma supervised learning akan mencoba menemukan batasan antara 10.000 gambar yang membaginya menjadi 2 bagian. Sehingga pada saat test ketika gambar baru, Misalkan gambar kucing datang sebagai input tanpa label apapun, algoritma akan secara otomatis memasukkannya kedalam keranjang kucing yang berlabel 0, Yang berarti algoritma tersebut mampu Memprediksi / Mengklasifikasi gambar tersebut sebagai gambar kucing.

**Contoh lain**, Kita adalah seorang real estate agent dan menghire trainee untuk membantu bisnis berkembang. Untuk membantu trainee memprediksi harga rumah, kita membuat data penjualan rumah untuk mereka. Dan yang paling penting adalah kita menulis harga final rumahnya pada data yang ada.

### Unsupervised Learning

Tidak seperti Supervised Learning, **Unsupervised Learning** tidak memiliki label pelatihan untuk sampel pelathian. Algoritma akan dirumuskan sedemikian rupa sehingga dapat menemukan struktur dan pola yang sesuai dalam data. Setelah pola yang konsisten ini menjadi jelas, titk data serupa dapat dikelompokkan bersama dan titik data yang berbeda akan berapa dalam kelompok yang berbeda. Ini sebagian besar digunakan untuk meproyeksikan data dimensi tinggi ke dalam dimensi rendah untuk tujuan visualisasi dan analisis.

**Tujuan utama-nya** adalah untuk memodelkan struktur dasar atau distribusi dalam data dengan tujuan untuk mempelajari data lebih jauh lagi. dengan kata lain, adalah menyimpulkan fungsi yang mendeskripsikan atau menjelaskan data. **Berbeda** dengan supervised learning disini tidak ada jawaban yang ***dibenarkan secara terarah*** dan tidak ada ***guru***.

### Reinforcement Learning
Adalah Jenis Machine Learning  yang memiliki agent (seperti robot) yang **belajar bagaimana berperilaku di lingkungan dengan mengambil tindakan dan mengukur hasilnya.** Jika tanggapan agent benar, ia akan mendapatkan reward point yang meningkatkan kepercayaan diri agen untuk mengambil lebih banyak tindakan semacam itu.

Dalam kehidupan sehari hari, Positive atau Negative feedback sesekali digunakan untuk reinforce perilaku. Hampir mirip seperti melatih anjing. Perilaku **reward-motivated** ini adalah kunci dalam Reinforcement Learning.

### Deep Learning
Adalah subkategori machine learning, Seperti yang telah dibahas sebelumnya, Ide AI terinspirasi dari otak manusia. Deep Learning terinspirasi oleh Artificial Neural Networks dan Artificial Neural Networks atau biasa dikenal ANN, ANN terinspirasi dari jaringan saraf biologis manusia.



Gambar diatas adalah Shallow Neural Network yang disebut Shallow Learning Network. Neural Network akan selalu memiliki :

- Input Layer   : It can be pixels of an image or a time series data
- Hidden Layer  : Commonly known as weights which are learned while the neural network is trained
- Output Layer  : Final layer mainly gives you a prediction of the input you fed into your network


Jika, Neural Network adalah approximation function dimana jaringan mencoba mempelajari parameter (bobot) di hidden layers yang bila dikalikan dengan input akan memberi kita predicted output mendekati output yang diinginkan.

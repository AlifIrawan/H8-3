### Histograms 
Histogram adalah cara untuk merepresentasikan distribusi frekuensi dari dataset numerik.

***Cara Kerja***


Mempartisi sumbu x menjadi beberapa bin, menetapkan setiap titik data dalam kumpulan data kita ke sebuah bin. dan kemudian menghitung jumlah titik data yang telah ditetapkan disetiap bin. Jadi sumbu y adalah frekuensi atau banyaknya titik data pada setiap bin. *Noted* kita dapat mengubah ukuran bin dan biasanya perlu menyesuaikan agar distribusinya ditampilkan dengan baik.

```python
df_can['2013'].head()
```

```python
# np.histogram returns 2 values
count, bin_edges = np.histogram(df_can['2013'])

print(count) # Frequency count
print('='*50)
print(bin_edges) # bin ranges, default = 10 bins
```

Secara default, *metode* ***histogram*** memecah dataset menjadi 10 bin. 

Dibawah ini merangkum kisaran bin dan distribusi frekuensi imigrasi di tahun 2013 :

- 178 Countries contributed between 0 to 3412.9 immigrants
- 11 Countries contributed between 3412.9 to 6825.8 immigrants
- 1 Countries contributed between 6825.8 to 10238.7 immigrants, and so on...

```python
df_can['2013'].plot(kind='hist', figsize=(8,5))

plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.xlabel('Number of Immigrants')
plt.ylabel('Number of Countries')

plt.show()
```

Pada plot diatas *Sumbu x* mennunjukkan kisaran populasi imigran dalam interval 3412,9 *Sumbu y* Mewakili jumlah negara yang kontribusi pada populasi tersebut.

***Perhatikan*** bahwa label sumbu x tidak cocok dengan ukuran bin, Hal ini dapat diperbaiki dengan memasukkan **xticks** yang berisi *Daftar ukuran bin*, Sebagai berikut :

```python
# 'bin_edge' is a list of bin intervals
count, bin_edges = np.histogram(df_can['2013'])

df_can['2013'].plot(kind='hist', figsize=(8,5), xticks=bin_edges)

plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.xlabel('Number of Immigrants')
plt.ylabel('Number of Countries')

plt.show()
```

***Let's Modified to look impact & aesthetic plot***

- Increase the bin size to 15 by passing in bins parameter
- set transparency to 60% by pasing in alpha parameter
- label the x-axis by passing in x-label parameter
- change the color of the plots by passing in color parameter

```python
# Let's get the x-tick values
count, bin_edges = np.histogram(df_DNS, 15)

# un-stacked histogram
df_DNS.plot(
            kind='hist',
            figsize=(10,6),
            bins=15,
            alpha=.6,
            xticks=bin_edges,
            color=['coral', 'darkslateblue', 'mediumseagreen']
            )

# Give explanation
plt.title("Histogram of Immigration from Denmark, Norway and Sweden from 1980 - 2013")
plt.xlabel("Number of Immigrants")
plt.ylabel("Number of Years")

# Result
plt.show()
```

Jika kita tidak ingin terlihat tumpang tindih, kita dapat menggunakan parameter **Stacked**. Sekaligus menyesuaikan label x min dan maks untuk menghilangkan celah ekstra di tepi plot. 

Kita bisa pass tuple (min, max) menggunakan parameter **xlim**.

```python
# Let's get the x-tick values
count, bin_edges = np.histogram(df_DNS, 15)

# First bin values is 31.0, adding buffer of 10 for aesthetic purposes
xmin = bin_edges[0] - 10

# Last bin values is 308.0, adding buffer of 10 for aesthetic purposes
xmax = bin_edges[-1] + 10

# un-stacked histogram
df_DNS.plot(
            kind='hist',
            figsize=(10,6),
            bins=15,
            alpha=.6,
            xticks=bin_edges,
            color=['coral', 'darkslateblue', 'mediumseagreen'],
            stacked=True,
            xlim=(xmin, xmax)
            )

# Give explanation
plt.title("Histogram of Immigration from Denmark, Norway and Sweden from 1980 - 2013")
plt.xlabel("Number of Immigrants")
plt.ylabel("Number of Years")

# Result
plt.show()
```

### Pie Chart

Adalah grafik circular yang menampilkan proporsi numerik dengan membagi lingkaran menjadi irisan proposional (Presentase). Parameter tambahan didalam plot adalah :

- ***autopct*** - is a string or function used to label the wedges with their numeric value. The label will be placed inside the wedge. if it is a format string, the label will be *fmt%pct*

- ***startangle*** - rotates the start of the pie chart by angle degrees counterclockwise from the x-axis

- ***shadow*** - Draws a shadow beneath the pie (to give a 3D feel)

```python
# Step 1: Get the data


# Group countries by continents and apply sum() function
df_continents = df_can.groupby('Continent', axis=0).sum()

# We can't use it further until we apply a function (ex: sum())
print(type(df_can.groupby('Continent', axis=-0)))

df_continents
```

```python
# Step 2: Plot data
df_continents['Total'].plot(
                            kind='pie', 
                            figsize=(5, 6), 
                            autopct='%1.1f%%',    # Add in percentages
                            startangle=90,        # Start angle 90 (Africa)
                            shadow=True           # Add shadow
                            )

plt.title('Immigrantion to Canada by Continent 1980 - 2013')
plt.axis('equal')

plt.show()
```

***Let's Modified to look impact & aesthetic plot***

- ***plt.legend()*** - Remove the text labels on the pie chart by passing in legend and add it as seperate legend

- ***pctdistance*** - Push out the percentages to sit just outside the pie chart

- ***explode*** - The pie chart to emphasize the lowest three continents (Africa, North America, & Latin America and Carribbean)

```python
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']

explode_list = [0, 0, .12, .1, 0, 0] # Ratio for each continent with which to offset each wedge

df_continents['Total'].plot(
                                kind = 'pie',
                                figsize = (15, 6),
                                autopct = '%1.1f%%',
                                startangle = 90,
                                shadow = True,
                                labels = None,           # Turn off labels on pie chart
                                pctdistance = 1.12,      # The ratio between the center of each pie slice and the start of the text generated by autopct
                                colors = color_list,     # add custom color
                                explode = explode_list   # 'explode' lowest 3 continents
                            )

# Scale the title up by 12% to match pctdistance
plt.title('Immigrantion to Canada by Continet from 1980 - 2013', y=1.12)
plt.axis('equal')

# add legend
plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

### SubPlots

Sering kali kita ingin memplot beberapa plot dalam figure yang sama. untuk itu kita dapat menggunakan *figure (overall canvas)* dan membaginya menjadi *subplots*. dalam subplot kita akan bekerja dengan ***artist layer*** daripada ***scripting layer***.

**Sintaks umum-nya:**

fig = plt.figure() # create figure

ax = fig.add_subplot(nrows, ncols, plot_number) # create subplots

Dimana:

- **nrows** & **ncols** are used to notionally split the figure into (**nrows** * **ncols**) sub-axes.
- **plot_number** is used to indentify the particular subplot that this function is to create within the notional grid. **plot_number** starts at 1, increments across rows first and has a maximum of **nrows** * **ncols** as shown below

```python
fig = plt.figure() # create figure

ax0 = fig.add_subplot(1,2,1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(122) # add subplot 1 (1 row, 2 columns, second plot)


# Subplot 1: Box plot
df_CI.plot(kind='box', figsize=(20, 6), vert=False, color='blue', ax=ax0) # add to subplot 1
ax0.set_title('Box Plots of Immigrants from China and India 1980 - 2013')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')


# Subplot 2: Line plot
df_CI.plot(kind='line', figsize=(20, 6), color=('red', 'blue'), ax=ax1) # add to subplot 2
ax1.set_title('Line Plots of Immigrants from China and India 1980 - 2013')
ax1.set_xlabel('Number of Immigrants')
ax1.set_ylabel('Years')

plt.show()
```

### Bubble Plots
Adalah variasi dari *Scatter plot* yang menampilkan 3 dimensi data (x,y,z)

**Let's Start off by analyzing the effect of Argentina's great depression**

Argentina mengalami great depression dari tahun 1998 - 2002, yang menyebabkan meluasnya pengangguran, kerusuhan jatuhnya pemerintahan dan gagal bayar hutang luar negri negara tersebut. Dalam hal pendapatan, lebih dari 50% orang argentina miskin, dan tujuh dari sepuluh anak argentina miskin pada krisis tahun 2002. Kita akan bandingkan data imigrasi dengan tetangga-nya Brazil

```python
# Step 1: Get the data
df_can_t = df_can[years].transpose()

# cast the years (the indeks) to type int
df_can_t.index = map(int, df_can_t.index)

# Lets label the index. this will automatically be the column name when we reset the index
df_can_t.index.name = 'Year' 

# Reset index to bring the Year in as a Column
df_can_t.reset_index(inplace=True)

# Result
df_can_t.head()
```

**Selanjutnya, Create the normalized data**

ada beberapa metode normalisasi dalam statistik. Masing masing dengan kegunaan-nya sendiri. Dalam kasus kali ini, akan menggunakan *feature scaling* untuk membawa semua nilai kedalam kisaran [0,1]



dimana Xold adalah nilai original. Xnew adalah nilai yang dinormalisasi. rumus-nya menetapkan nilai maksimal dalam kumpulan data ke 1, dan menetapkan nilai min ke 0. titik data lainnya diskalakan ke nilai antara 0-1

```python
# Normalize Brazil data
norm_brazil = (df_can_t['Brazil'] - df_can_t['Brazil'].min()) / (df_can_t['Brazil'].max() - df_can_t['Brazil'].min())

# Normalize Argetina data
norm_argetina = (df_can_t['Argentina'] - df_can_t['Argentina'].min()) / (df_can_t['Argentina'].max() - df_can_t['Argentina'].min())
```

**Plot Data from Dataset**

```python
ax0 = df_can_t.plot(kind='scatter', 
                    x='Year', 
                    y='Brazil', 
                    figsize=(14, 8), 
                    alpha=.5, 
                    color='green', 
                    s=norm_brazil * 2000 + 10, 
                    xlim=(1975, 2015)
                   )

ax1 = df_can_t.plot(kind='scatter', 
                    x='Year', 
                    y='Argentina', 
                    alpha=.5, 
                    color='blue', 
                    s=norm_argetina * 2000 + 10, 
                    ax = ax0
                   )

ax0.set_ylabel('Number of immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```

Ukuran bubble sesuai dengan besarnya populasi yang berimigrasi untuk tahun itu, dibandingkan dengan data 1980 - 2013. *Semakin besar bubble berarti semakin banyak imigran ditahun tersebut.*
